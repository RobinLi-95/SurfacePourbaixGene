2. Find the path for the Vasp Calculations

```python
PATH = "/Users/libingxin/Project/SurfacePourbaix/Mg_10m10/Vacuum/onsurface_adsorption/"
# PATH = "/Volumes/bl4518/projects/corrosion/live/Bingxin/ELF_Paper/OH_adsorption/OH_ad"
os.chdir(PATH)

# Read the directory of the calculated Mg surface
dir = subprocess.getoutput("ls -p | grep /")
dir = re.split('\W+',dir)

repo = []
for i in range(20):
    if 'Mg_' in dir[i]:
        repo.append(dir[i])
```

```python
Surf = []

for dir in repo:
    os.chdir(dir)
    outputdata = vasprd("POSCAR")
    outputdata.SurfaceDetermine()

    surf = {"Name":str(dir), "ChemSym":outputdata.chemicalsymbol, 
            "Number_Mg":outputdata.N_Mg, "Number_O":outputdata.N_O, "Number_H":outputdata.N_H, 
            "G_constant": outputdata.G0, "k_index":outputdata.k_factor}

    Surf.append(surf)
    os.chdir('..')
```

* show the surface information with the Pandas Data Framwork

```python
df = pd.DataFrame(Surf)
df
```

3. Determine the most stable surface phase at different pH and potential

```python
test_surf = pourbaixGen(Surf)
test_surf.TwoDPhaseDetermination(0,14,-3,1,500,500)
```
